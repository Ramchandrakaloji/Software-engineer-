```python

import unittest

class PrimeChecker:

    def is_prime(self, n):
        # Check if the number is prime
        if n <= 1:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True

    def generate_primes(self, limit):
        # Generate list of prime numbers up to the given limit
        primes = []
        for num in range(2, limit + 1):
            if self.is_prime(num):
                primes.append(num)
        return primes

class TestPrimeChecker(unittest.TestCase):

    def setUp(self):
        # Initialize the PrimeChecker instance
        self.checker = PrimeChecker()

    def test_primes(self):
        # Test cases for prime numbers
        self.assertTrue(self.checker.is_prime(2))
        self.assertTrue(self.checker.is_prime(17))
        self.assertTrue(self.checker.is_prime(97))

    def test_non_primes(self):
        # Test cases for non-prime numbers
        self.assertFalse(self.checker.is_prime(1))
        self.assertFalse(self.checker.is_prime(10))
        self.assertFalse(self.checker.is_prime(100))

    def test_prime_list(self):
        # Test case for generating prime numbers up to a limit
        self.assertEqual(self.checker.generate_primes(10), [2, 3, 5, 7])

# Run the tests
unittest.main(argv=[''], verbosity=2, exit=False)
```
